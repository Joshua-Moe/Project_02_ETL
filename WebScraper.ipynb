```python
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd
import re
import argparse
import logging
from pprint import pprint
```

```python
html_file = "Resources/List_of_United_States_state_birds.html"

HTMLFileToBeOpened = open(html_file, "r", encoding="utf8")
contents = HTMLFileToBeOpened.read()
soup = BeautifulSoup(contents, 'lxml')
HTMLFileToBeOpened.close()
```

```python
search_results = soup.find_all('table', attrs={"class":"wikitable sortable"})
```

```python
search_results[0]
```

```python
state_birds = search_results[0].find_all('tr')
```

```python
state_birds[0]
```

```python
for ti in state_birds[1]:
    print(ti.text.strip())
```

```python
# Lists to store the State and Bird Names
States = []
Birds = []

# Looping through the state bird rows
for row in state_birds[1:]:
    
    state_name = row.find('th').text.strip()
    
    ti = row.find_all('td')
    state_bird = ti[0].text.strip()
    
    # Append to lists
    States.append(state_name)
    Birds.append(state_bird)

    
```

```python
States
```

```python
Birds
```

```python
state_bird_df = pd.DataFrame({'State':States, 'State Bird':Birds})
state_bird_df.head()
```

```python
state_bird_df.to_csv('State_Birds.csv', index=False)
```
